# Prototype
Here we will prototype our code for cleaning the data.

## Objectives
Remove and columns and rows that are empty or NULL. Update or aggregate columns into clearer column names. Change any values to be more simple, clean, and clear.

# Code

Read the raw CSV file

```python
import pandas as pd
from pathlib import Path

csv_path = Path("../Data/Raw/geothermal_wells_ca.csv")
df = pd.read_csv(csv_path)
df.head()
```

Inspect the size of the data

```python
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

Print the amount of values each column has

```python
for col, count in df.count().items():
    print(f"  '{col}': {count} values")
```

```python
print(df.isnull().sum())
```

```python
empty_columns = df.columns[
    df.apply(lambda col: col.isna().all() or (col.fillna(0) == 0).all())
].tolist()
print(empty_columns)
```

## Data Cleaning

Drop the missing columns

```python
df.drop(columns=empty_columns, inplace=True)
print(f"After dropping empty columns, remaining columns: {df.columns.tolist()}")
```

Print the size after removing the empty columns

```python
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("Removed 10 columns that were entirely empty or zero-filled.")
```

Remove rows with missing coordinates

```python

missing_coordinate = df["Lat83"].isna() | df["Long83"].isna()


print(f"Missing Lat83 or Long83:  {missing_coordinate.sum()} rows")
print(f"Total rows:               {len(df)} rows")
print(f"Percentage missing coordinates:  {missing_coordinate.sum() / len(df) * 100:.2f}%")
```

Drop the rows with missing cordinate and print the amount of rows left

```python
df.dropna(subset=["Lat83", "Long83"], inplace=True)
print(f"Rows after dropping missing coordinates: {len(df)}")
```

Print the final size of the data

```python
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```
